```python
mode = widgets.ToggleButtons(
    options=['Kids Mode 🧒', 'Young Adult Mode 🎓'],
    description='Select Mode:'
)

display(mode)
```

```python
kids_challenges = [
    "Turn off the tap while brushing your teeth 🚰",
    "Pick up 3 pieces of litter 🌱",
    "Use a reusable lunch container 🍱",
    "Take a 5-minute shower 🚿"
]

adult_challenges = [
    "Avoid single-use plastic today ♻",
    "Carry a reusable water bottle 💧",
    "Refuse a plastic straw 🌊",
    "Track your water usage today 📊"
]

print("\n🎯 Today's Challenge:")

# Choose challenge list based on mode
if mode.value == 'Kids Mode 🧒':
    challenge_list = kids_challenges
else:
    challenge_list = adult_challenges

challenge_index = today.day % len(challenge_list)
daily_challenge = challenge_list[challenge_index]

print(daily_challenge)

# Streak tracker
if "streak" not in globals():
    streak = 0

button = widgets.Button(description="I completed today's challenge!")
output = widgets.Output()

def on_button_clicked(b):
    global streak
    streak += 1
    with output:
        output.clear_output()
        print(f"🔥 Your streak: {streak} days")

button.on_click(on_button_clicked)

display(button, output)
```

```python
question = "Which resource is most scarce in South Africa?"
options = ["Coal", "Water", "Gold", "Wind"]

quiz = widgets.RadioButtons(
    options=options,
    description='Choose:',
)

submit = widgets.Button(description="Submit Answer")
quiz_output = widgets.Output()

def check_answer(b):
    with quiz_output:
        quiz_output.clear_output()
        if quiz.value == "Water":
            print("✅ Correct! Water scarcity is a major issue in SA.")
        else:
            print("❌ Not quite. Try again tomorrow!")

submit.on_click(check_answer)

print("\n🧠 Quick Quiz:")
print(question)

display(quiz, submit, quiz_output)
```

```python
if "days_completed" not in globals():
    days_completed = set()

def complete_day(day):
    days_completed.add(day)
    print(f"🔥 Progress: {len(days_completed)}/7 days completed")

print("\n📅 7-Day Climate Challenge")

for i in range(1, 8):
    btn = widgets.Button(description=f"Complete Day {i}")
    btn.on_click(lambda b, day=i: complete_day(day))
    display(btn)

```

```python
provinces = [
    "Gauteng",
    "Western Cape",
    "KwaZulu-Natal",
    "Eastern Cape",
    "Limpopo"
]

province_selector = widgets.Dropdown(
    options=provinces,
    description="Select Province:"
)

display(province_selector)
```

```python
province_facts = {
    "Gauteng": "Gauteng faces high water demand due to dense population and aging infrastructure.",
    "Western Cape": "The Western Cape experienced a severe drought between 2017–2018, nearly reaching 'Day Zero' in Cape Town.",
    "KwaZulu-Natal": "Flooding and infrastructure damage in KwaZulu-Natal have impacted water systems in recent years.",
    "Eastern Cape": "Parts of the Eastern Cape experience recurring drought conditions and water supply disruptions.",
    "Limpopo": "Limpopo is semi-arid and heavily dependent on rainfall and dam storage for water security."
}
```

```python
fact_output = widgets.Output()

def show_province_fact(change):
    with fact_output:
        fact_output.clear_output()
        selected = province_selector.value
        print("💧 Province Water Fact:")
        print(province_facts[selected])

province_selector.observe(show_province_fact, names="value")

display(fact_output)
```

```python
print("\n🌊 Water Impact Calculator")

people_input = widgets.IntSlider(
    value=10,
    min=1,
    max=1000,
    step=1,
    description='Number of People:'
)

days_input = widgets.IntSlider(
    value=7,
    min=1,
    max=30,
    step=1,
    description='Days:'
)

calc_button = widgets.Button(description="Calculate Water Saved")
calc_output = widgets.Output()

def calculate_water(b):
    with calc_output:
        calc_output.clear_output()

        people = people_input.value
        days = days_input.value

        # Assume 45 litres saved per 5-minute shower
        water_saved = people * days * 45

        print(f"💧 Total Water Saved: {water_saved} litres")
        print("That’s enough to supply multiple households for days.")

calc_button.on_click(calculate_water)

display(people_input, days_input, calc_button, calc_output)
```
